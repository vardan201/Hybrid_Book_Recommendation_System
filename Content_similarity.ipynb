```python
import re

def clean_tags(text):
    if isinstance(text, str):
        tags = text.split()
        tags = [re.sub(r'^-+|-+$', '', tag) for tag in tags]        # remove dashes
        tags = [tag.lower() for tag in tags if len(tag) > 2]        # keep meaningful tags
        tags = [re.sub(r'[^a-zA-Z\-]', '', tag) for tag in tags]    # remove special characters
        return ' '.join(tags)
    return ''

books['tags'] = books['tags'].apply(clean_tags)
```

```python
books['tags'][2]
```

```python
def refine_tags(tag_str):
    if not isinstance(tag_str, str):
        return ''

    tags = tag_str.lower().split()
    cleaned_tags = []

    for tag in tags:
        tag = tag.strip('-')  # remove leading/trailing dashes
        if len(tag) < 3 or tag in {'d', 'read', 'default', 'my'}:
            continue  # remove noise

        # normalize common variations
        if 'sci' in tag and 'fi' in tag:
            tag = 'scifi'
        elif tag in ['favourite', 'favorites', 'favourites', 'favs', 'faves']:
            tag = 'favorite'
        elif tag.startswith('read-in'):
            tag = 'read'
        elif tag == 'distopian':
            tag = 'dystopian'

        cleaned_tags.append(tag)

    return ' '.join(set(cleaned_tags))  # remove duplicates
books['tags'] = books['tags'].apply(refine_tags)
```

```python
!pip install unidecode
import re
import unidecode

def clean_author(author_str):
    author_str = author_str.lower()                            # lowercase
    author_str = unidecode.unidecode(author_str)               # remove accents
    author_str = re.sub(r'[^\w\s]', '', author_str)            # remove punctuation
    author_str = re.sub(r'\s+', ' ', author_str).strip()       # remove extra spaces
    return author_str
```

```python
books['authors'] = books['authors'].apply(clean_author)
```

```python
import re

def clean_title(title):
    title = title.lower()
    title = re.sub(r'\([^)]*\)', '', title)  # remove anything in parentheses
    title = re.sub(r'[^a-z0-9\s]', '', title)  # remove special characters
    title = re.sub(r'\s+', ' ', title).strip()  # clean extra spaces
    return title
books['title'] = books['title'].apply(clean_title)
```

```python
books['content'] = books['title'] + ' ' + books['authors'] + ' ' + books['tags']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(books['content'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(books.index, index=books['title']).drop_duplicates()
```

```python
indices
```

```python
import re

def get_recommendations(title, cosine_sim=cosine_sim):
    # Clean input title (remove punctuation and lower it)
    title = title.lower()
    title = re.sub(r'\([^)]*\)', '', title)  # remove series info like (Twilight #2)
    title = re.sub(r'[^a-z0-9\s]', '', title).strip()

    # Find the index of the book in your DataFrame
    idx = indices.get(title)
    if idx is None:
        return "❌ Book not found. Please check the title."

    # Get similarity scores with all other books
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort scores from highest to lowest
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Pick top 10 most similar books (excluding the input book itself)
    sim_scores = sim_scores[1:11]

    # Get book indices
    book_indices = [i[0] for i in sim_scores]

    # Return the recommended book titles
    return books['title'].iloc[book_indices]
```

```python
get_recommendations("the first world war")
```
